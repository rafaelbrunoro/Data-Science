```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importa a biblioteca de data
from datetime import datetime

#Cria uma figura para plotagem do gráfico
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

#Lê o arquivo com os dados a serem utilizados
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
#Associa a coluna SPX à variável spx
spx = data['SPX']

#Plota no gráfico os dados de spx no eixo criado
spx.plot(ax=ax, style='k-')

#cria um dicionário para plotagem das anotações eixos (x, y)
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

#Navega pelos dados da crise de 2008/2009
for date, label in crisis_data:
    #Adiciona uma anotação ao rótulo recuperado
    ax.annotate(label, 
                #Coloca a seta a uma distância de 75 unidades para a data atual recuperada
                xy=(date, spx.asof(date) + 75),
                #Coloca o texto a uma distância de 225 unidades para a data atual recuperada
                xytext=(date, spx.asof(date) + 225),
                #Cria uma seta na cor preta com largura da seta de 4 unidades, 
                #largura do corpo de 2 unidades e 
                #tamanho da seta de 4 unidades
                arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
                #Alinha o texto à esquerda e no topo
                horizontalalignment='left', verticalalignment='top')

#Limita a exibição dos dados entre 2007 e 2011
ax.set_xlim(['1/1/2007', '1/1/2011'])
#Limita o índice do S&P 500 entre 600 e 1800 pontos
ax.set_ylim([600, 1800])

#Adiciona um título a plotagem
ax.set_title('Datas importantes na crise de 2008-2009 financial')

#Salva em um arquivo o gráfico gerado 
plt.savefig('graphics.png', dpi=400, bbox_inches='tight')
```
